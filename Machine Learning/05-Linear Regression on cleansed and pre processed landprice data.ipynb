```python
saleprice_corr=data.corr()['SalePrice']
saleprice_corr
```

```python
saleprice_corr[abs(saleprice_corr)>0.1].index
```

```python
req_columns=saleprice_corr[abs(saleprice_corr)>0.1].index
```

```python
datafinal=data[req_columns]
```

```python
datafinal.head()
```

```python
y = datafinal['SalePrice']
X= datafinal.drop('SalePrice',axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
type(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train,y_train)
```

```python
m  = lin_reg.coef_
print(m)
```

```python
c = lin_reg.intercept_
c
```

```python
ypred_test=lin_reg.predict(X_test) #for testing
ypred_train=lin_reg.predict(X_train) #for training
```

```python
ypred_test
```

```python
ypred_train
```

```python
from sklearn.metrics import mean_squared_error #calculating MSE
MSE_test=mean_squared_error(ypred_test,y_test)
print("MSE for testing is",MSE_test)
```

```python
MSE_train=mean_squared_error(ypred_train,y_train)
print("MSE for training is",MSE_train)
```

```python
((MSE_train-MSE_test)/MSE_train)*100
```

```python
#calculating rsquared value
lin_reg.score(X_train,y_train) #R2 VALUE FOR TRAINING
```

```python
lin_reg.score(X_test,y_test) #R2 VALUE FOR TESTING
```
